```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sal_data = pd.read_csv("/content/Dataset09-Employee-salary-prediction.csv")
sal_data.head()
```

```python
sal_data.shape
```

```python
sal_data.columns
```

```python
sal_data.columns = ['Age', 'Gender', 'Degree', 'Job_Title', 'Experince_years', 'Salary']
sal_data.head()
```

```python
sal_data[sal_data.duplicated()]
sal_data1=sal_data.drop_duplicates(keep = 'first')
sal_data1.shape
```

```python
sal_data1.isnull().sum()
sal_data1.dropna(how = 'any', inplace=True)
```

```python
sal_data1.shape
```

```python
sal_data1.Age.plot(kind = 'box')
```

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder = LabelEncoder()
sal_data1['Gender_Encode'] = Label_Encoder.fit_transform(sal_data1['Gender'])
sal_data1['Degree_Encode'] = Label_Encoder.fit_transform(sal_data1['Degree'])
sal_data1['Job_Title_Encode'] = Label_Encoder.fit_transform(sal_data1['Job_Title'])
sal_data1.head()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
sal_data1['Age_scaled'] = std_scaler.fit_transform(sal_data1[['Age']])
sal_data1['Experince_years_scaled'] = std_scaler.fit_transform(sal_data1[['Experince_years']])
sal_data1.head()
```

```python
x = sal_data1[['Age_scaled', 'Gender_Encode', 'Degree_Encode', 'Job_Title_Encode', 'Experince_years_scaled']]
y = sal_data1['Salary']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
Linear_regression_model = LinearRegression()
Linear_regression_model.fit(x_train, y_train)
```

```python
y_pred_lr = Linear_regression_model.predict(x_test)
y_pred_lr
```

```python
df = pd.DataFrame({'y_Actual': y_test, 'y_Predicted': y_pred_lr})
df['Error'] = df['y_Actual'] - df['y_Predicted']
df['abs_error'] = abs(df['Error'])
df
```

```python
Mean_absolute_Error = df['abs_error'].mean()
Mean_absolute_Error
```

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
r2_score(y_test, y_pred_lr)
print(f"Accuracy of the model is {round(r2_score(y_test, y_pred_lr),4)*100}%")
```

```python
round(mean_absolute_error(y_test, y_pred_lr),2)
print(f"Mean Absolute Error of the model is {round(mean_absolute_error(y_test, y_pred_lr),2)}")
```

```python
mse = round(mean_squared_error(y_test, y_pred_lr),2)
print(f"Mean Squared Error of the model is {round(mean_squared_error(y_test, y_pred_lr),2)}")
```

```python
print("Root Mean Squared Error of the model is", mse**(0.5))
```
